```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('figure', figsize=(10, 10))
```

```python
x = np.arange(-4, 4, 0.1)
y = np.arange(-4, 4, 0.1)
X, Y = np.meshgrid(x, y) # Matrix of X, Y containing the x and y coordinate of every point in the grid
```

```python
# Equation
Z = X*Y*(2 - X - 2*Y)
p = plt.contour(X, Y, Z, np.arange(-1, 1, 0.5))
plt.clabel(p, inline=1, fontsize=10)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
```
